```python
def TopTenPlot(column):
    global imdb
    imdb[column].value_counts().sort_values(ascending=False)[:10].plot(kind="bar", figsize=(20,6), edgecolor="black")
    plt.xticks(rotation=0)
    plt.title("Top Ten {}".format(column))
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.show()
```

```python
TopTenPlot("Director")
```

```python
TopTenPlot("Actor 1")
```

```python
TopTenPlot("Actor 2")
```

```python
TopTenPlot("Actor 3")
```
